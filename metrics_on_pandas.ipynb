## преобразование из object в datetime

```python
df_merged['publishedDate'] = df_merged['publishedDate'].astype(str)
```

```python
df_merged['publishedDate'] = df_merged.publishedDate.str.split(r'-\s*').str[0]
```

```python
df_splitted = df_merged.publishedDate.str.split(r'-\s*').str[0]
df_splitted_wrong = df_splitted.str.find('-')
df_splitted_wrong.value_counts()
```

```python
df_merged['publishedDate'] = df_merged['publishedDate'].str.replace('*', '')
```

```python
df_merged['publishedDate'] = df_merged['publishedDate'].str.replace('?', '0')
```

```python
df_merged['publishedDate'] = df_merged['publishedDate'].fillna('')
```

```python
df_merged['publishedDate'] = pd.to_datetime(df_merged['publishedDate'])
```

## 2 способ

```python
def color_cells(val):
    if val == 'True':
        color = 'limegreen'
    elif val == 'False':
        color = 'darksalmon'
    else:
        color = 'slategrey'
    return 'background-color: %s' % color
```

```python
df_merged['90_dates'] = df_merged.publishedDate.str.find('1990') == 0.0
```

```python
def color_rows(row):
    if row['90_dates'] == 'True':
        return ['background-color: limegreen'] * len(row)
    else:
        return ['background-color: darksalmon'] * len(row)
```

```python
df_merge_style = df_merged.style.apply(color_rows, axis=1)
```

```python
print(df_merge_style.render())
```
